```python
youtube = data[['Youtuber', 'Subscribers', 'Video Views', 'Uploads', 
                'Category', 'Country', 'Abbreviation', 'Lowest Monthly Earnings', 
                'Highest Monthly Earnings', 'Gross Tertiary Education Enrollment (%)', 
                'Unemployment Rate', 'Population', 'Urban Population', 'Created Year']] #, 'Latitude', 'Longitude'
```

```python
#There are special characters in the data columns 'Youtuber' and 'Title' that need to be processed.
#remove special characters
import re
pattern = r'[^a-zA-Z0-9\s.,!?&\'-]'
#Youtuber feature
youtube['Youtuber'] = youtube['Youtuber'].apply(lambda x: re.sub(pattern, '', x)) 
youtube['Youtuber'] = youtube['Youtuber'].str.strip() 
youtube['Youtuber'] = youtube['Youtuber'].str.lstrip('- ') 
```

```python
#check special characters in Youtuber feature
filter_youtuber_rows = youtube['Youtuber'].str.contains(pattern, regex = True)
filter_youtuber_rows_results = filter_youtuber_rows[filter_youtuber_rows == True]
print("Youtuber's special characters:",filter_youtuber_rows_results)
```

```python
#Remove data null in column 'Created Year'
youtube = youtube.dropna(subset= ['Created Year'])
#youtube['Created Year'] = youtube['Created Year'].astype('object')
# Convert 'Created Year' to object type using .loc
youtube.loc[:, 'Created Year'] = youtube['Created Year'].astype('object')
youtube.info()
```

```python
# Check for blanks in the 'Youtuber' column
blanks_data_youtuber = youtube[youtube['Youtuber'] == '']
blanks_data_youtuber
```

```python
# remove rows where the 'Youtuber' column is blank using boolean indexing. Afterward, 
# reset the index of the DataFrame and check again for blanks to confirm the removal. 
# The reset_index(drop=True) part ensures that the index is reset without keeping the old index as a new column.
# Remove rows with blank 'Youtuber' and reset the index
youtube = youtube[youtube['Youtuber'] != ''].reset_index(drop=True)

# Check for blanks in the 'Youtuber' column after removal
blanks_data_youtuber = youtube[youtube['Youtuber'] == '']
blanks_data_youtuber
```

```python
#Check duplicate data
youtube.duplicated().sum()
```

```python
#Check zero data video views
zero_views = youtube[youtube['Video Views'] == 0]
zero_views
```

#### There are a number of YouTube channels which have 0 video views, these appear to be YouTube topics & not valid channels, for the purpose of this analysis these rows will be removed

```python
# Remove rows with zero video views and reset the index
youtube = youtube.drop(zero_views.index)
youtube.reset_index(drop=True, inplace=True)

# Display rows with zero video views after removal (should be an empty DataFrame)
print("\nRows with zero video views after removal:")
youtube[youtube['Video Views'] == 0]
```

```python
youtube.shape
```

```python
#Check null/nan data
youtube.isnull().sum()
```

```python
#There are lots of missing value in the category variables
#We will filter all missing to 'Other'

#Select category columns
categorical_variables = youtube.select_dtypes(include= 'O').columns

#Replace object datatype column missing values with 'Other'
youtube[categorical_variables] = youtube[categorical_variables].fillna('Other')
```

```python
#check missing value
youtube.isna().sum()
```

```python
youtube.info()
```

```python
# Remove rows where key columns have missing values
youtube_not_null = youtube.dropna(subset=['Gross Tertiary Education Enrollment (%)', 'Unemployment Rate', 'Population', 'Urban Population'])

# Calculate and print the mean of Gross Tertiary Education Enrollment
mean_edu = round(youtube_not_null['Gross Tertiary Education Enrollment (%)'].mean(), 1)
print('Mean Gross Tertiary Education Enrollment:', mean_edu)

# Calculate and print the mean of Unemployment Rate
mean_unemp = round(youtube_not_null['Unemployment Rate'].mean(), 1)
print('Mean Unemployment Rate:', mean_unemp)

# Calculate and print the mean of Population
mean_popu = round(youtube_not_null['Population'].mean(), 0)
print('Mean Population:', mean_popu)

# Calculate and print the mean of Urban Population
mean_urban = round(youtube_not_null['Urban Population'].mean(), 0)
print('Mean Urban Population:', mean_urban)
```

```python
# There are lots of missing value in the numeric variables
# Replace missing values in numeric columns with mean values
# Replace missing values in 'Gross Tertiary Education Enrollment (%)' with the previously calculated mean
youtube['Gross Tertiary Education Enrollment (%)'] = youtube['Gross Tertiary Education Enrollment (%)'].fillna(mean_edu)

# Replace missing values in 'Unemployment Rate' with the previously calculated mean
youtube['Unemployment Rate'] = youtube['Unemployment Rate'].fillna(mean_unemp)

# Replace missing values in 'Population' with the previously calculated mean
youtube['Population'] = youtube['Population'].fillna(mean_popu)

# Replace missing values in 'Urban Population' with the previously calculated mean
youtube['Urban Population'] = youtube['Urban Population'].fillna(mean_urban)
```

```python
youtube.info()
```

```python
# Convert specified float columns to integer data type

# Convert 'Video Views' column to int64
youtube['Video Views'] = youtube['Video Views'].astype('int64')

# Convert 'Created Year' column to int64
youtube['Created Year'] = youtube['Created Year'].astype('int64')

# Convert 'Population' column to int64
youtube['Population'] = youtube['Population'].astype('int64')

# Convert 'Urban Population' column to int64
youtube['Urban Population'] = youtube['Urban Population'].astype('int64')

youtube.info()
```

```python
# Sort the DataFrame by the 'Subscribers' column in descending order

# Use the sort_values method to sort the DataFrame by 'Subscribers' in descending order
youtube = youtube.sort_values(by='Subscribers', ascending=False)

# Reset the index after sorting to reflect the new order
youtube = youtube.reset_index(drop=True)

# Display the sorted DataFrame
youtube
```

#### Top 10 YouTubers in terms of subscribers and the Top 10 YouTubers with the highest views?¶

```python
# Select columns to display in the output
select_columns = ['Youtuber', 'Subscribers']

# Filter the DataFrame to get the top 10 Youtuber channels with the most subscribers
top_10_subs = youtube[select_columns].head(10)

# Convert 'Subscribers' to millions and add a new column 'Subscribers (million)'
top_10_subs['Subscribers (million)'] = (top_10_subs['Subscribers'] / 1000000).astype('int')

# Keep only the relevant columns in the final DataFrame
top_10_subs = top_10_subs[['Youtuber', 'Subscribers (million)']]

# Display the final DataFrame containing the top 10 Youtuber channels with the most subscribers
top_10_subs
```

```python
# Select columns to display in the output
select_columns = ['Youtuber', 'Video Views']

# Filter the DataFrame to get the top 10 Youtuber channels with the most video views
youtube_videoviews = youtube[select_columns]
top_10_views = youtube_videoviews.sort_values(by='Video Views', ascending=False).head(10)

# Convert 'Video Views' to billions and add a new column 'Video Views (billion)'
top_10_views['Video Views (billion)'] = (top_10_views['Video Views'] / 1000000000).astype('int')

# Keep only the relevant columns and sort the final DataFrame
top_10_views = top_10_views[['Youtuber', 'Video Views (billion)']].sort_values(by='Video Views (billion)', ascending=False).reset_index(drop=True)

# Display the final DataFrame containing the top 10 Youtuber channels with the most video views
top_10_views
```

```python
# Create a figure and axis
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Bar chart for top 10 subscribers
top_10_subs.plot(kind='bar', x='Youtuber', y='Subscribers (million)', ax=ax1, color='blue')
ax1.set_title('Top 10 Youtuber Channels by Subscribers')
ax1.set_ylabel('Subscribers (Million)')

# Bar chart for top 10 video views
top_10_views.plot(kind='bar', x='Youtuber', y='Video Views (billion)', ax=ax2, color='green')
ax2.set_title('Top 10 Youtuber Channels by Video Views')
ax2.set_ylabel('Video Views (Billion)')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

**- From the analysis of the top 10 channels with the highest subscribers and video views, we observe that 7 out of the top 10 channels with the highest subscribers also appear in the top 10 channels with the highest video views.

**-The rankings show a consistent pattern, with 'T-Series' maintaining the top position in both categories.

**- We can also abserve that there is not a significant change in the rankings when transitioning from subscribers to video views. Most the of channels maintain similar positions in both lists, indicating a correlation between subscriber count and video views.

**Correlation Analysis:**
- An interesting question arises: Is there a correlation between the number of subscribers and views? Let's delve deeper into the relationship between these two variables in the next section.


#### Does a channel having more subcribers imply that it will have more views?

```python
# Numerical columns for correlation analysis
numerical_columns = ['Subscribers', 'Video Views', 'Uploads', 
                      'Lowest Monthly Earnings', 'Highest Monthly Earnings', 
                      'Gross Tertiary Education Enrollment (%)', 'Unemployment Rate', 
                      'Population', 'Urban Population', 'Created Year']

# a new DataFrame containing only the numerical columns
youtube_numerical = youtube[numerical_columns]

# correlation matrix calculation
correlation_matrix = youtube_numerical.corr()

# Display the correlation matrix
correlation_matrix

```

```python
# Setting the style of the visualization
sns.set(style="white")

# Calculating the correlation matrix
correlation_matrix = youtube_numerical.corr()

# Setting up the matplotlib figure
plt.figure(figsize=(10, 8))

# Creating a heatmap using Seaborn
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=.5)

# Set the title of the heatmap
plt.title("Correlation Matrix Heatmap")

# Show the plot
plt.show()
```

#### There is a strong positive correlation of approximately 0.85 between Subscribers and Video Views. 
#### This suggests that channels with a higher number of subscribers tend to have more video views, and vice versa.

#### The correlation coefficients between Subscribers or Video Views and Uploads are relatively low, indicating, a weak correlation. #### Here are the specific values from the correlation matrix:

##### The correlation between Subscribers and Uploads is approximately 0.09.
##### The correlation between Video Views and Uploads is approximately 0.16.
#### These values suggest that there is a weak positive correlation between the number of uploads and both the number of subscribers and video views. 
#### However, the correlations are not strong, and other factors may have a more significant influence on the number of subscribers and views.
